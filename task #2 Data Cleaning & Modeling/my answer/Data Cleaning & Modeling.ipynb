## ***Data Cleaning***

```python
# in Content Table remove NULL Column
Content.drop('URL', axis='columns', inplace=True)
```

```python
# in Content Table remove NULL Row Contain
Reactions.dropna(inplace=True)
```

```python
# irrelevant observations
Profile = Profile[(Profile.Interests.str[1] != '0') & (Profile.Age > 10)]
# Profile.Interests.unique()
# Profile.Age.unique()
User
```

```python
# irrelevant observations
import re 
regex_patern = r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+'
User = User[User.Email.str.match(regex_patern) == True]
User.head()
```

```python
# remove noise data
l = ["sbnnz", "thbhn", "ouafw", "dcppv", "xtzqw","qxugb"]
Session =  Session[~Session.Device.isin(l)]
Session.Device.value_counts()
```

```python
Session = Session[Session.Duration != 0]
```

```python
Session[Session.Duration == 0]
```

```python
Content.Category = Content.Category.str.strip('"').str.strip().str.lower()
# Content.Category.value_counts()
```

### ***Data modeling***

#### **JOIN Tables**

```python
join_1 = Profile.merge(Content, on='User ID')
```

```python
join_1.head(5)
```

```python
join_2 = join_1.merge(Reactions, on='Content ID', 
                      suffixes=("_Profile", "_Content"))
```

```python
join_2.head(5)
```

```python
first_table = join_2.merge(ReactionTypes, left_on='Type_Content', right_on='Type', 
                      suffixes=("_Reactions", "_ReactionTypes"))
```

```python
first_table
```

```python
join_3 = Location.merge(User, on='User ID')
```

```python
join_3
```

```python
second_table =  join_3.merge(Session, on='User ID')
```

```python
second_table
```

```python
Final_data_set = first_table.merge(second_table, left_on = 'User ID_Profile', right_on = 'User ID')
```
